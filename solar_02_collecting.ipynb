```python
import re

def make_flat(data, year_2016=True):

    columns = data.columns

    data_cloud = pd.Series()
    for column in columns:
        data_cloud = data_cloud.append(data[column])
    data_cloud.dropna(inplace=True)
    
    data_cal = data_wando['일시'].copy()

    data_cal = data_cal.apply(lambda x : "".join(re.findall('\d', x)))

    if year_2016:
        data_cal = data_cal[data_cal < np.str(20170000)]
    else:
        data_cal = data_cal[data_cal > np.str(20170000)]
    
    data_result = pd.DataFrame()

    if year_2016:
        data_result['data'] = data_cal.values
        data_result['cloud'] = data_cloud.values
    else:
        data_result['data'] = data_cal.values
        data_result['cloud'] = data_cloud.values[:198]
        
    return data_result
```
