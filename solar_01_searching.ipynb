# [태양광발전과 태양광 예측: 최신 기술 동향](http://www.ndsl.kr/ndsl/search/detail/report/reportSearchResultDetail.do?cn=KOSEN0000000761226)

1. 인트로
    - 초록 : 화석연료의 과도한 사용으로 인한 에너지 고갈과 환경오염의 문제가 발생되었다. 이를 해결하기 위해서 청정한 에너지인 태양광발전이 급속하게 개발되었고, 이는 전력망에 연결되고 있다. 하지만 태양광발전은 **시간적 및 공간적 변수와 이런 변수에 의한 불확실성 때문에 예측이 쉽지 않다.** 본 보고서에서는 이런 예측의 다양한 방법론과 정확도에 관한 최근의 결과 및 동향에 대해 서술하고 있다. 예측 기간에 따라서 최적의 결과를 보이는 방법론이 있다. 하루 내의 단기 예측(intra-day forecast: 0~6시간 전)에는 하늘 및 위성사진이 포함된 지속성 모델(persistence model)이 우수한 성능을 발휘한다. 하루 전의 예측(day ahead forecast)에는 수치 날씨 예측(numerical weather prediction) 모델과 사후 처리에 의해 최적의 결과가 얻어진다. 그리고 예측 정확도의 다양한 지표들이 개발되었고, 근평균제곱오류(RMSE: root mean square error) 등이 많이 사용된다. 그리고 예측 시스템 평가의 중요한 기준은 다양한 운영자들의 설문을 바탕으로 결정되고 만들어진다.
    - 태양광발전 시스템의 이슈 : 변수, 불확실성
        - 태양광발전 시스템의 높은 침투에 대한 두 가지 큰 과제는 변수와 불확실성이다. 즉 태양 광발전 출력은 시간적 변수에 노출되어 있다.
    - 태양광발전 용도에 따라 다른 예측의 형태가 필요하다. 
        - 단일 태양광 발전 시스템
        - 다수 태양광 발전 시스템 : 다수의 태양광발전의 예측이 필요할 때는 업스케일링 방법이 사용된다. 업스케일링된 물리학적 접근법은 Lorenz et al.에 의해 적용되 었다.

2. 예측 방법

    - 기준 1. 물리학적(physical) 혹은 통계학적(statistical)
        - 물리학적 접근은 태양광발전을 예측 하기 위한 태양광과 태양광발전 모델을 사용 한다. 태양광 발전량의 예측은 관련된 날씨의 변수와 태양광발전 시스템의 특성의 함수로 이루어진다.
            - NWP(수치 날씨 예측 모델)
                - 수치 날씨 예측 모델(NWP)은 초기 조건 으로부터 며칠의 대기 진화를 예측하는 역학을 기본으로 한다.
                - 예측의 개선은 종종 예측을 보정하는 동 안 측정된 자료와 비교함으로써 이루어진다. 이러한 접근법은 종종 MOS(Model Output Statistics)이라고 한다.
        - 반면에 통계학적 접근은 태양광이나 태양광발전 모델에 의존하지 않고, 모델을 만들기 위해서 과거의 자료에 주로 의존한다.
        > 통계학적 접근법과 물리학적 접근법을 비교했을 때 통계학적 접근법이 물리학적 접근법보다 조금 뛰어났다. 그러나 실제적으로 이 두 접근법은 혼용되어 사용될 수 있다.

    - 기준 2. 시간 - DA(day ahead forecast), HA(hour ahead forecast)
        - 하루 전의 예측[day ahead(DA) forecast]은 태양광발전이 시작되는 전날 5시 30분에 제출. 예측되는 날의 18.5 에서 42.5 시간 전에 대해 제공.
        - 한 시간 전의 예측[hour ahead(HA) forecast]은 태양광발전이 되는 매시간의 105분 전에 제출. CAISO HA 예측은 1.75 에서 8.75시간 전에 대한 것.
        > 태양광 예측에서는 예측 기간에 따라 다른 방법론이 선호된다.
        - 0~6 시간 전의 태양광 및 태양광발 전의 예측(Intra-day forecast) - HA
            - 토털 하늘 사진
            - 위성 구름 이동 벡터 접근
            - 확률 학습 기법
                - 매우 짧은 기간(하루 내의 2~3 시간 전) 의 예측을 위해서는 외부의 입력값이 없는 확률 학습 기법이 정확도와 쉬운 설치 측면에서 매우 경쟁력이 있다.
        - 6~24 시간 전의 태양광 및 태양광발 전의 예측
            - 태양광과 태양광발전 예측 중 가장 많이 사용되는 것은 시간당 출력의 하루 전 예측(day-ahead forecasting)

    - 기준 3. 지리적 범위 - Point 예측과 area 예측
        - 시스템의 일부의 자료를 이용하여 주어 진영역에설치된모든전력시스템의전체 출력을 예측하는 것을 일반적으로 업스케일링이라고 한다.
        - 업스케일링 기술은 계산을 단순화시키고, 주 어진 영역에 설치된 모든 태양광발전 시스템에 대한 상세한 특징에 대한 필요성을 감소시킨다.
        - 업스케일링의 요건
            - 첫째, 일부분은 전력 출력의 거동이 태양광발전 시스템의 전체의 거동을 대표하는 것으로 선택되어야 한다. 
            - 둘째, 전체 시스템의 전력출력을 얻기 위해서 일부 시스템의 전력출력을 조정하는 것이다.
      
3. 예측의 정확도에 영향을 미치는 요인
    - 태양광과 태양광발전의 예측은 기후 와 날씨의 조건에 크게 의존한다.
        - 지리적인 의존성은 날씨의 상황(하늘청명계수 “kT”로 흔히 대변되는 구 름의 상태) 태양의 고도각, 지형 상태의 요인들로 설명될 수 있다.
        - 태양 고도각의 분포는 예측의 지리적 의존 성뿐만 아니라, 계절의 의존성을 야기시킨다.
    - 단일장소 혹은 지역적 예측
        - 지리적 면적 크기와 관측소 혹은 시스템의 개수 모두 예측 오류를 감소시키는데 기여
    - 예측 기간
        - 하루 이틀 전의 예측범위에서는 "수치날씨 예측 모델"이 가장 좋다. 
        - 6시간보다 짧은 예측범위에서는 "기상데이터 기반 통계모델"이 좋다.
        - 예측의 정확도는 예측 기간이 증가함에 따라 떨어지고, 과거의 자료에 기반을 둔 지속성 예측방법의 경우 더욱 급격하게 떨어진다.
        > 최대 여섯 시간 전 데이터까지 활용하여 발전량 예측해야
 
4. 태양광 예측 조사 결과(예측 현황)
    - 대부분의 모델들은 6~72시간전의 예측에 대해 다루는 수치날씨 예측모델이다.
    - 순수한 통계모델은 거의 없다.

# [서포트 벡터 회귀를 이용한 24시간 후의 태양광 발전량 예측](http://www.riss.kr/search/detail/DetailView.do?p_mat_type=be54d9b8bc7cdb09&control_no=481438ebdca2a3e0ffe0bdc3ef48d419#redirect)

최근 화석 연료의 고갈과 환경 문제로 인해 신재생에너지의 중요성이 증가하고 있다. 그 중에서도 특히, 설치가 용이하고 무한한 자원을 가진 태양광 에너지가 각 광받고 있다. 하지만 태양광 발전은 기상에 많은 영향을 받기 때문에 발전량이 일 정하지 않아 예측에 어려움이 있다. 24시간 앞의 태양광 발전량을 예측하는 것은 한국 전력거래소의 전력시스템과 스마트그리드 시스템에 안정적으로 연계되어 효율 적인 에너지관리를 가능하게 할 수 있다. 본 연구에서 24시간 앞의 태양광 발전량 예측 모델을 수립하기 위하여, 2013년부터 2014년까지의 발전량, 기상예보 및 기상 실측 데이터를 수집하였다. 그리고 기계학습 알고리즘인 서포트 벡터 회귀를 이용 하여 일사량을 예측하는 모델을 수립하였고, 예측된 일사량으로부터 최종적인 24시 간 앞의 태양광 발전량을 예측하는 모델을 수립하였다. 예측 모델 수립과정에서 그 리드 서치 방법을 통해 적절한 파라미터를 찾고 변수 후진제거 방법을 통해 모델 성능에 영향을 미치지 않은 변수를 제거하여 최종 예측모델을 수립하였다. 그리고 기존 연구에서 사용된 인공신경망 모형들과의 예측성능을 비교하였다. 또한, 발전량 예측을 저해하는 요인들의 영향을 분석하였고 구름두께의 불확실성을 고려하여 모 델의 일반적인 성능을 도출하였다.

1. 태양광 발전량에 영향을 미치는 요인
    - 기상요인과 지리요인 그리고 설비요인으로 구분할 수 있다. 
    - 기상요인
        - 일사량 : 태양에서 오는 빛의 복사(輻射)
        - 온도 : 태양전지는 반도체이기 때문에 고온에서는 태양광 발전 효율이 떨어지기 때문
        - 일조시간, 운량, 상대습도, 풍향, 풍속, 강우량, 강 설량, 황사, 미세먼지 : 그 외 기상요소 
    - 지리요인
        - 위도, 경도 및 해발높이
    - 설비요인
        - 고정형, 고정가변형, 추적형으로 구분되는 발전시스템의 태양전지 모듈 경사각의 차이
        - 축전지 및 전력변환장치의 효율

2. 데이터 : 기상예보 정보, 기상실측 정보, 태양고도 정보
    - 기상예보 정보 : 하늘상태, 강수확률, 습도, 온도, 풍향, 풍속
    - 기상실측 정보 : 일사량, 운량, 자외선B(태양광 중에서 파장이 280~315nm인 빛)
    - 태양고도 정보 : 태양고도는 월별로 차이를 보이기 때문에 월마다 차이를 보이는 발전량에 영향을 미칠 것
    > 지리요인, 설비요인은 반영하지 않음
    
3. 예측기간 : 장기예측, 단기예측
    - 단기예측은 태양광 발전소를 효율적으로 운영 하거나 다른 전력계통과 안정적인 연계를 가능하게하기 위해 필요하다. 
    - 초단기예측, 24시간 예측
        - 초단기예측
            - 초단기 발전량을 예측하는 방법으로는 실측 기상정보를 활용하여 모델을 만들고, 실시간으로 제공되는 기상정보를 입력하여 발전량을 예측한 방법
            - 태양광 패널이 설치되어 있는 지점의 하늘 상태를 촬영한 이미지를 통해 하늘의 청명도를 산출하여 이동평균 및 선형회 귀 모형으로 예측한 방법
        - 24시간 예측
            - 인공신경망 모델과 SVR(support vector regression) 모델로 발전량 예측

4. 예측모델
    - 아래 데이터를 활용하여 예측모델 수립
    - 기상예보(하늘상태, 강수확률, 습도, 온도, 풍향, 풍속) 데이터
    - 실측기상(운량, 자외선B)데이터
    - 태양고도 데이터
